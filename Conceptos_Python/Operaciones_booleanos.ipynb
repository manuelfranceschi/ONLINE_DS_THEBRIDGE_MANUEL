## Operaciones con booleanos
[al indice](#Contenidos)  

Todas las operaciones que realizabamos en el apartado anterior devolvan un tipo de dato concreto: un booleano. `True` o `False`. Pero cmo haras si se tienen que cumplir 3 condiciones, o solo una de esas tres, o que no se cumplan 5 condiciones?
Para este tipo de operaciones recurrimos al [*lgebra de Boole*](https://es.wikipedia.org/wiki/%C3%81lgebra_de_Boole#:~:text=El%20%C3%A1lgebra%20de%20Boole%2C%20tambi%C3%A9n,que%20esquematiza%20las%20operaciones%20l%C3%B3gicas.). Se trata de una rama del lgebra que se utiliza en electrnica, pero que tiene un sin fin de aplicaciones, no solo tncicas, sino aplicables a la vida cotidiana. Estas matemticas pueden llegar a ser muy complejas an utilizando unicamente dos valores: `True` y `False`. Las operaciones ms comunes son **AND, OR, NOR**.
En las siguientes tablas tienes todos los posibles resultados de las puertas AND, OR, NOR, dependiendo de sus inputs.



Puede parecer complejo pero a efectos prcticos, y sin meternos con otro tipo de puertas lgicas, te recomiendo seguir estas reglas:
* **AND**: Se tienen que cumplir ambas condiciones para que sea un `True`
* **OR**: Basta que se cumpla al menos una condicion para que sea `True`
* **NOR**: Lo contrario de lo que haya

Veamos un ejemplo prctico para aclarar estos conceptos. Imaginemos que queremos comprar un ordenador, pero nos cuesta decidirnos. Eso s, tenemos claras las siguentes condiciones a la hora de elegir
* La RAM me vale que tenga 16, 32 o 64 GB
* En cuanto al procesador y disco duro, la combinacin que mejor me viene es un i3 con 500GB de disco.
* Precio: que no pase de los 800 

```python
# Primer ordenador
ram_1 = 32
process_1 = "i5"
disco_1 = 500
precio_1 = 850

# Segundo ordenador
ram_2 = 8
process_2 = "i5"
disco_2 = 500
precio_2 = 600

# Tercer ordenador
ram_3 = 32
process_3 = "i3"
disco_3 = 500
precio_3 = 780
```

Veamos cmo implemento esto mediante operaciones booleanas

```python
# Primero, calculamos el valor de estas condiciones por separado
# para el primer ordenador
cond_ram_1 = (ram_1 == 16 or ram_1 == 32 or ram_1 == 64) #OR -> Me vale con cualquier resultado a True
cond_process_1 = (process_1 == "i3" and disco_1 == 500) # AND -> obliga a que todas las condiciones deban ser True
cond_precio_1 = (precio_1 < 800)
cond_total_1 = cond_ram_1 and cond_process_1 and cond_precio_1
print("Todas las condiciones:",cond_total_1, cond_ram_1, cond_process_1, cond_precio_1)
```

El primer ordenador cumple el requisito de ram, pero no los de precio y procesador/disco. Veamos los otros dos si los cumplen

```python
# Calculamos las condiciones para los otros ordenadores
# para el segundo ordenador
cond_ram_2 = (ram_2 == 16 or ram_2 == 32 or ram_2== 64) #OR -> Me vale con cualquier resultado a True
cond_process_2 = (process_2 == "i3" and disco_2 == 500) # AND -> obliga a que todas las condiciones deban ser True
cond_precio_2 = (precio_2 < 800)
cond_total_2 = cond_ram_2 and cond_process_2 and cond_precio_2
print("Todas las condiciones:",cond_total_2, cond_ram_2, cond_process_2, cond_precio_2)

# para el tercer ordenador
cond_ram_3 = (ram_3 == 16 or ram_3 == 32 or ram_3 == 64) #OR -> Me vale con cualquier resultado a True
cond_process_3 = (process_3 == "i3" and disco_3 == 500) # AND -> obliga a que todas las condiciones deban ser True
cond_precio_3 = (precio_3 < 800)
cond_total_3 = cond_ram_3 and cond_process_3 and cond_precio_3
print("Todas las condiciones:",cond_total_3, cond_ram_3, cond_process_3, cond_precio_3)
```

Bingo! El tercer ordenador cumple todas las condiciones para ser mi futura compra. Vers en prximos notebooks que esto se puede hacer todava ms sencillo mediante bucles y funciones.
